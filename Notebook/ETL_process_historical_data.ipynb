```python
# Create matches dataset from players resource table
match_df = players_df[["match_id", "team", "opponent", "best_of", "map_1", "map_2", "map_3", "date"]]
match_df = match_df.drop_duplicates("match_id").reset_index(drop=True)
match_df.count()
```

```python
# Prepare match dataframe to fit the SQL database
match_df["date"] = pd.to_datetime(match_df["date"])
match_df = match_df.rename(columns={"team": "team_1", "opponent": "team_2"})
match_df
```

```python
# Split results dataset to get results for each team
columns_team_1 = ["match_id", "team_1", "_map", "result_1", "map_winner", "starting_ct", "ct_1", "t_1", "rank_1", "map_wins_1", "match_winner"]
columns_team_2 = ["match_id", "team_2", "_map", "result_2", "map_winner", "starting_ct", "ct_2", "t_2", "rank_2", "map_wins_2", "match_winner"]
results_1_df = results_df[columns_team_1]
results_1_df
```

```python
results_1_df["map_winner"] = results_1_df["map_winner"].apply(lambda x: "W" if (x == 1) else "L")
results_1_df["starting_ct"] = results_1_df["starting_ct"].apply(lambda x: "Yes" if (x == 1) else "No")
results_1_df["match_winner"] = results_1_df["match_winner"].apply(lambda x: "W" if (x == 1) else "L")

results_1_df
```

```python
results_2_df = results_df[columns_team_2]
results_2_df
```

```python
results_2_df["map_winner"] = results_2_df["map_winner"].apply(lambda x: "W" if (x == 2) else "L")
results_2_df["starting_ct"] = results_2_df["starting_ct"].apply(lambda x: "Yes" if (x == 2) else "No")
results_2_df["match_winner"] = results_2_df["match_winner"].apply(lambda x: "W" if (x == 2) else "L")

results_2_df
```

```python
results_1_df.rename(columns = {
                                "team_1": "team", 
                                "_map": "map", 
                                "result_1":"rounds_won",
                                "ct_1": "rounds_won_ct",
                                "t_1": "rounds_won_t",
                                "rank_1": "rank",
                                "map_wins_1": "map_wins"
}, inplace=True) 
results_1_df.head()
```

```python
results_2_df = results_2_df.rename(columns = {
                                "team_2": "team", 
                                "_map": "map", 
                                "result_2":"rounds_won",
                                "ct_2": "rounds_won_ct",
                                "t_2": "rounds_won_t",
                                "rank_2": "rank",
                                "map_wins_2": "map_wins"
}) 
results_2_df.head()
```

```python
team_results_df = pd.concat([results_1_df, results_2_df], ignore_index=True)
team_results_df
```
