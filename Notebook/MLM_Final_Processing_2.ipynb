```python
# Only match results that didnt end in overtime(more than 30 rounds)

cleaned_results = results_df.loc[(results_df['result_1'] <= 16) & (results_df['result_2'] <= 16)]
```

```python
cleaned_results = cleaned_results[['date', 'team_1', 'team_2', '_map',
       'map_winner',
       'match_id', 'total_rounds']].copy()
```

```python
cleaned_results
```

```python
# Seperating Results by team
team_1_results = cleaned_results[['date', 'team_1', '_map',
                                  'map_winner','match_id', 'total_rounds']].copy()

team_2_results = cleaned_results[['date', 'team_2', '_map',
                                  'map_winner','match_id', 'total_rounds']].copy()
```

```python
# New result team headers
team_headers = ['date', 'team', '_map', 'map_winner','match_id', 'total_rounds']
```

```python
# Chnaging the map_winner value from 1 or 2 to W or L
team_2_results.loc[team_2_results.map_winner == 2, 'map_winner'] = 'W'
team_2_results.loc[team_2_results.map_winner == 1, 'map_winner'] = 'L'

team_1_results.loc[team_1_results.map_winner == 2, 'map_winner'] = 'L'
team_1_results.loc[team_1_results.map_winner == 1, 'map_winner'] = 'W'
```

```python
# Renaming team columns
team_1_results.columns = team_headers
team_2_results.columns = team_headers
```

```python
player_results_rev = pd.merge(players_revised, cleaned_results, on=['match_id','_map','date'])
```

```python
# Averages for player performances (There are outliers in the dataset)
player_results_rev.describe()
```

```python
# Creating New features based on player statistics and rounds played
```

```python
KPR = player_results_rev['kills']/ player_results_rev['total_rounds']
player_results_rev['KPR'] = KPR
```

```python
SPR = (player_results_rev['total_rounds'] - player_results_rev['deaths'])/player_results_rev['total_rounds']
player_results_rev['SPR'] = SPR
```

```python
DPR = player_results_rev['deaths']/player_results_rev['total_rounds']
player_results_rev['DPR'] = DPR
```

```python
APR = player_results_rev['assists']/player_results_rev['total_rounds']
player_results_rev['APR'] = APR
```

```python
organised_player_results = player_results_rev.sort_values(by=['player_id','date'])
organised_player_results
```

```python
# Establishing player averages to build player ratings off of
Average_KPR = organised_player_results.groupby(['player_id']).KPR.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
Average_KPR
```

```python
organised_player_results.loc[:,'Average_KPR'] = Average_KPR
```

```python
Average_SPR = organised_player_results.groupby(['player_id']).SPR.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
Average_SPR
```

```python
organised_player_results.loc[:,'Average_SPR'] = Average_SPR
```

```python
Average_DPR = organised_player_results.groupby(['player_id']).DPR.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
Average_DPR
```

```python
organised_player_results.loc[:,'Average_DPR'] = Average_DPR
```

```python
Average_APR = organised_player_results.groupby(['player_id']).APR.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
Average_APR
```

```python
organised_player_results.loc[:,'Average_APR'] = Average_APR
```

```python
Average_KAST = organised_player_results.groupby(['player_id']).kast.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
Average_KAST
```

```python
organised_player_results.loc[:,'Average_KAST'] = Average_KAST
```

```python
Average_ADR = organised_player_results.groupby(['player_id']).adr.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
Average_ADR
```

```python
organised_player_results.loc[:,'Average_ADR'] = Average_ADR
```

```python
# Building player stat ratings per map
KPR_rating = organised_player_results['KPR']/organised_player_results['Average_KPR']
KPR_rating
```

```python
organised_player_results.loc[:,'KPR_rating'] = KPR_rating
```

```python
DPR_rating = organised_player_results['DPR']/organised_player_results['Average_DPR']
DPR_rating
```

```python
organised_player_results.loc[:,'DPR_rating'] = DPR_rating
```

```python
SPR_rating = organised_player_results['SPR']/organised_player_results['Average_SPR']
SPR_rating
```

```python
organised_player_results.loc[:,'SPR_rating'] = SPR_rating
```

```python
APR_rating = organised_player_results['APR']/organised_player_results['Average_APR']
APR_rating
```

```python
organised_player_results.loc[:,'APR_rating'] = APR_rating
```

```python
KAST_rating = organised_player_results['kast']/organised_player_results['Average_KAST']
KAST_rating
```

```python
organised_player_results.loc[:,'KAST_rating'] = KAST_rating
```
