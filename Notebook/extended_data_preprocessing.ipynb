```python
import pandas as pd
```

```python
players_revised = pd.read_csv('players_extended_stats.csv')
players_addition = pd.read_csv('players_extended_stats_rating_addition.csv')
```

```python
results = pd.read_csv('results.csv')
```

```python
players_revised = players_revised.drop(['Unnamed: 0'], axis=1)
players_addition = players_addition.drop(['Unnamed: 0'], axis=1)
```

```python
players_revised.sort_values(['match_id', 'team'])
```

```python
players_addition.sort_values(['match_id', 'team'])
```

```python
players_revised = pd.merge(players_revised,players_addition,how='left', on=['date', 'player_name', 'team', 'opponent', 'player_id', 'match_id',
       'event_id', 'best_of', 'map', 'kills', 'assists', 'deaths', 'hs',
       'flash_assists', 'kast', 'kddiff', 'adr', 'fkdiff', 'rating'])
```

```python
players_addition.columns
```

```python
players_revised.columns
```

```python
map_teamStats = {    'kills': 'sum',
    'assists': 'sum',
    'deaths': 'sum',
    'hs': 'sum',
    'flash_assists': 'sum',
    'kast': 'mean',
    'kddiff': 'sum',
    'adr': 'mean',
    'fkdiff': 'sum',
    'rating': 'mean',
    'Kill_Avg_Per_Map': 'sum', 
    'Assits_Avg_Per_Map': 'sum', 
    'Deaths_Avg_Per_Map': 'sum',
    'HS_Avg_Per_Map':'sum', 
    'Assists_Avg_Per_Map': 'sum', 
    'Kast_Avg_Per_Map': 'mean',
    'Kddiff_Avg_Per_Map': 'sum', 
    'ADR_Avg_Per_Map':'mean', 
    'FKdiff_Avg_Per_Map': 'sum',
    'Kill_Avg_Overall': 'sum', 
    'Assits_Avg_Overall': 'sum', 
    'Deaths_Avg_Overall': 'sum',
    'HS_Avg_Overall': 'sum', 
    'Assists_Avg_Overall': 'sum', 
    'Kast_Avg_Overall': 'mean',
    'Kddiff_Avg_Overall': 'sum', 
    'ADR_Avg_Overall': 'mean', 
    'FKdiff_Avg_Overall': 'sum',
    'Rating_Avg_Per_Map':'mean', 
    'Rating_Avg_Overall':'mean'
                }
playersBothTeams_df = players_revised.groupby(['match_id', 'team','opponent', 'event_id', 'best_of', 'map']).agg(map_teamStats)
playersBothTeams_df
```

```python
playersFinal_df = pd.DataFrame(playersBothTeams_df)
```

```python
playersFinal_df.reset_index(inplace=True)
```

```python
playersFinal_df
```

```python
team_1 = playersFinal_df.loc[1::2,:]
team_1
```

```python
team_2 = playersFinal_df.loc[::2,:]
team_2
```

```python
team_1.isnull().sum()
```

```python
team_1_headers = ['match_id', 'team_1', 'opponent', 'event_id', 'best_of', 'map', '1_kills',
       '1_assists', '1_deaths', '1_hs', '1_flash_assists', '1_kast', '1_kddiff', '1_adr',
       '1_fkdiff', '1_rating', '1_Kill_Avg_Per_Map', '1_Assits_Avg_Per_Map',
       '1_Deaths_Avg_Per_Map', '1_HS_Avg_Per_Map', '1_Assists_Avg_Per_Map',
       '1_Kast_Avg_Per_Map', '1_Kddiff_Avg_Per_Map', '1_ADR_Avg_Per_Map',
       '1_FKdiff_Avg_Per_Map', '1_Kill_Avg_Overall', '1_Assits_Avg_Overall',
       '1_Deaths_Avg_Overall', '1_HS_Avg_Overall', '1_Assists_Avg_Overall',
       '1_Kast_Avg_Overall', '1_Kddiff_Avg_Overall', '1_ADR_Avg_Overall',
       '1_FKdiff_Avg_Overall','1_Rating_Avg_Per_Map', '1_Rating_Avg_Overall']

team_2_headers = ['match_id', 'team_2', 'opponent', 'event_id', 'best_of', 'map', '2_kills',
       '2_assists', '2_deaths', '2_hs', '2_flash_assists', '2_kast', '2_kddiff', '2_adr',
       '2_fkdiff', '2_rating', '2_Kill_Avg_Per_Map', '2_Assits_Avg_Per_Map',
       '2_Deaths_Avg_Per_Map', '2_HS_Avg_Per_Map', '2_Assists_Avg_Per_Map',
       '2_Kast_Avg_Per_Map', '2_Kddiff_Avg_Per_Map', '2_ADR_Avg_Per_Map',
       '2_FKdiff_Avg_Per_Map', '2_Kill_Avg_Overall', '2_Assits_Avg_Overall',
       '2_Deaths_Avg_Overall', '2_HS_Avg_Overall', '2_Assists_Avg_Overall',
       '2_Kast_Avg_Overall', '2_Kddiff_Avg_Overall', '2_ADR_Avg_Overall',
       '2_FKdiff_Avg_Overall','2_Rating_Avg_Per_Map', '2_Rating_Avg_Overall']
```

```python
team_1.columns = (team_1_headers)
team_2.columns = (team_2_headers)
```
