```python
# Obtaining our sample dataset
matches = ['2338858', '2339384', '2338862', '2339378']
results_sample_df = results_G2_df[results_G2_df.match_id.isin(matches)]
results_sample_df
```

```python
# Obtaining teams list
df = results_sample_df[["team_1", "team_2"]].drop_duplicates()
df1 = df["team_1"].drop_duplicates()
df2 = df[df["team_2"] != 'G2']
df2 = df2["team_2"].drop_duplicates()
df3 = pd.concat([df1, df2])
teams = df3.to_list()
teams
```
