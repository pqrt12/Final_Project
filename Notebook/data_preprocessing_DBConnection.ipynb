```python
players_df = pd.read_csv('players.csv')
```

```python
map_1_df = players_df[[
    'date',
    'player_name',
    'team',
    'opponent', 
    'player_id', 
    'match_id',
    'event_id',
    'best_of',
    'map_1',
    'm1_kills',
    'm1_assists',
    'm1_deaths',
    'm1_hs',
    'm1_flash_assists',
    'm1_kast',
    'm1_kddiff',
    'm1_adr',
    'm1_fkdiff',
    'm1_rating']].copy()
map_1_df
```

```python
map_1_df = map_1_df.dropna()
```

```python
map_2_df = players_df[[
    'date',
    'player_name',
    'team',
    'opponent', 
    'player_id', 
    'match_id',
    'event_id',
    'best_of',
    'map_2',
    'm2_kills',
    'm2_assists',
    'm2_deaths',
    'm2_hs',
    'm2_flash_assists',
    'm2_kast',
    'm2_kddiff',
    'm2_adr',
    'm2_fkdiff',
    'm2_rating']].copy()
map_2_df
```

```python
map_2_df = map_2_df.dropna()
```

```python
map_3_df = players_df[[
    'date',
    'player_name',
    'team',
    'opponent', 
    'player_id', 
    'match_id',
    'event_id',
    'best_of',
    'map_3',
    'm3_kills',
    'm3_assists',
    'm3_deaths',
    'm3_hs',
    'm3_flash_assists',
    'm3_kast',
    'm3_kddiff',
    'm3_adr',
    'm3_fkdiff',
    'm3_rating']].copy()
map_3_df
```

```python
map_3_df = map_3_df.dropna()
```

```python
map_3_df.columns
```

```python
rename_column_headers = ['date',
    'player_name',
    'team', 
    'opponent',
    'player_id',
    'match_id',
    'event_id',
    'best_of',
    'map',
    'kills',
    'assists',
    'deaths',
    'hs',
    'flash_assists',
    'kast',
    'kddiff',
    'adr',
    'fkdiff',
    'rating']

# Change column headers
map_1_df.columns = (rename_column_headers)
map_2_df.columns = (rename_column_headers)
map_3_df.columns = (rename_column_headers)
```

```python
frames = [map_1_df, map_2_df, map_3_df]

revised_players = pd.concat(frames)
```

```python
player_running_averages = revised_players.sort_values(['player_id','map','date'])
```

```python
player_running_averages
```

```python
pk = player_running_averages[['player_id','match_id','date','map','kills','assists','deaths','hs','flash_assists','kast','kddiff','adr','fkdiff','rating']]
```

```python
pk = pk.set_index('player_id')
```

```python
pk
```

```python
# Indivdual player stat cumulative average for every map
pk_cum_avg_kills = pk.groupby(['player_id','map']).kills.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_assists = pk.groupby(['player_id','map']).assists.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_deaths = pk.groupby(['player_id','map']).deaths.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_hs = pk.groupby(['player_id','map']).hs.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_flash_assists = pk.groupby(['player_id','map']).flash_assists.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_kast = pk.groupby(['player_id','map']).kast.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_kddiff = pk.groupby(['player_id','map']).kddiff.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_adr = pk.groupby(['player_id','map']).adr.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_cum_avg_fkdiff = pk.groupby(['player_id','map']).fkdiff.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
```

```python
pk_cum_avg_rating = pk.groupby(['player_id','map']).rating.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
```

```python
pk.loc[:,'Rating_Avg_Per_Map'] = pk_cum_avg_rating
```

```python
pk.loc[:,'Kill_Avg_Per_Map'] = pk_cum_avg_kills
pk.loc[:,'Assits_Avg_Per_Map'] = pk_cum_avg_assists
pk.loc[:,'Deaths_Avg_Per_Map'] = pk_cum_avg_deaths
pk.loc[:,'HS_Avg_Per_Map'] = pk_cum_avg_hs
pk.loc[:,'Assists_Avg_Per_Map'] = pk_cum_avg_flash_assists
pk.loc[:,'Kast_Avg_Per_Map'] = pk_cum_avg_kast
pk.loc[:,'Kddiff_Avg_Per_Map'] = pk_cum_avg_kddiff
pk.loc[:,'ADR_Avg_Per_Map'] = pk_cum_avg_adr
pk.loc[:,'FKdiff_Avg_Per_Map'] = pk_cum_avg_fkdiff
```

```python
player_running_averages = pd.merge(player_running_averages,pk, on=['player_id','match_id','date','map','kills','assists','deaths','hs','flash_assists','kast','kddiff','adr','fkdiff','rating'])
```

```python
player_running_averages = player_running_averages.sort_values(['player_id','date'])
```

```python
pk = player_running_averages[['player_id','match_id','date','map','kills','assists','deaths','hs','flash_assists','kast','kddiff','adr','fkdiff','rating']]
```

```python
pk = pk.set_index('player_id')
```

```python
pk_overall_avg_rating = pk.groupby(['player_id']).rating.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
```

```python
# Individual players stat cumulative average from earliest date to most current
pk_overall_avg_kills = pk.groupby(['player_id']).kills.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_assists = pk.groupby(['player_id']).assists.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_deaths = pk.groupby(['player_id']).deaths.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_hs = pk.groupby(['player_id']).hs.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_flash_assists = pk.groupby(['player_id']).flash_assists.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_kast = pk.groupby(['player_id']).kast.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_kddiff = pk.groupby(['player_id']).kddiff.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_adr = pk.groupby(['player_id']).adr.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
pk_overall_avg_fkdiff = pk.groupby(['player_id']).fkdiff.apply(lambda x : x.expanding(min_periods=1).mean().fillna(x))
```

```python
pk.loc[:,'Rating_Avg_Overall'] = pk_overall_avg_rating
```

```python
pk.loc[:,'Kill_Avg_Overall'] = pk_overall_avg_kills
pk.loc[:,'Assits_Avg_Overall'] = pk_overall_avg_assists
pk.loc[:,'Deaths_Avg_Overall'] = pk_overall_avg_deaths
pk.loc[:,'HS_Avg_Overall'] = pk_overall_avg_hs
pk.loc[:,'Assists_Avg_Overall'] = pk_overall_avg_flash_assists
pk.loc[:,'Kast_Avg_Overall'] = pk_overall_avg_kast
pk.loc[:,'Kddiff_Avg_Overall'] = pk_overall_avg_kddiff
pk.loc[:,'ADR_Avg_Overall'] = pk_overall_avg_adr
pk.loc[:,'FKdiff_Avg_Overall'] = pk_overall_avg_fkdiff
```

```python
player_running_averages = pd.merge(player_running_averages,pk, on=['player_id','match_id','date','map','kills','assists','deaths','hs','flash_assists','kast','kddiff','adr','fkdiff','rating'])
```
